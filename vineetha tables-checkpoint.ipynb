```python
plt.plot([2,4,6,4],color="r")
plt.ylabel("number")
plt.xlabel('indices')
plt.title('my plot')
plt.grid()
plt.show()
```

```python
l=["r","b","y"]
for i in l: 
    plt.plot([2,4,6,4],color=i)
    plt.ylabel("number")
    plt.xlabel('indices')
    plt.title('my plot')
    plt.grid()
    plt.show()
```

```python
l=["r","b","y"]
k=[".","v","1","0"]
for i in l:
    for j in k:
        for p in range (10,15):
            for n in 'd':
                plt.plot([2,4,6,4],color=i ,marker=j,markersize=p)
                plt.ylabel("number")
                plt.xlabel('indices')
                plt.title('my plot')
                plt.grid()
                plt.show()
```

```python
l=["r","b","y"]
k=[".","v","1","0"]
d=["-","--","-.",":"]
for i in l:
    for j in k:
        for p in range (10,15):
            for n in d:
                plt.plot([2,4,6,4],color=i ,marker=j,markersize=p,ls=n)
                plt.ylabel("number")
                plt.xlabel('indices')
                plt.title('my plot')
                plt.grid()
                plt.show()
```

```python
l=["r","b","y"]
k=[".","v","1","0"]
d=["-","--","-.",":"]
x=range(2,22,2)
y=range(3,33,3)
for i in l:
    for j in k:
        for p in range (10,15):
            for n in d:
                plt.plot(x,y,color=i ,marker=j,markersize=p,ls=n)
                plt.ylabel("number")
                plt.xlabel('indices')
                plt.title('my plot')
                plt.grid()
                plt.show()
```

```python
l=["r","b","y"]
k=[".","v","1","0"]
d=["-","--","-.",":"]
x=range(2,22,2)
y=range(3,33,3)
x1= np.array(x)*2
y1= np.array(y)*3
x2= np.array(x)**2
y2= np.array(y)**3

for i in l:
    for j in k:
        for p in range (10,15):
            for n in d:
                plt.plot(x,y,color=i,marker=j,markersize=p,ls=n)
                plt.plot(x1,y1,color=i,marker=j,markersize=p,ls=n)
                plt.plot(x2,y2,color=i,marker=j,markersize=p,ls=n)
                plt.ylabel("number")
                plt.xlabel('indices')
                plt.title('my plot')
                plt.grid()
                plt.show()
```

```python
import numpy as np
t=np.arange(0.,5.,0.2)
for i in range (1,16):
    plt.plot(t,t**2,'b--',label='^2.2')
    plt.plot(t,t**2.2,'rs',label='^2.2')
    plt.plot(t,t**22.5,'g^',label='^2.5')
    plt.grid()
    plt.xlabel("x axis")
    plt.ylabel("y axis")
    plt.legend()
    plt.show()
```
